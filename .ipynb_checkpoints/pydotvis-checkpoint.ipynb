```python
def text_to_json(text):
    """
    Convert indented text into a JSON-like dictionary.

    Parameters:
    - text (str): The indented text to convert. Each line represents a node, and the indentation level
                  of the line represents the node's level in the hierarchy.

    Returns:
    - dict: The indented text converted to a JSON-like dictionary. Each key is a node, and its value is a
            dictionary of its child nodes.
    """
    
    # Split the text into lines
    lines = text.split('\n')
    
    # Initialize an empty dictionary to build up the final hierarchy
    hierarchy = {}

    # Initialize a stack to keep track of the current node at each indentation level
    # Each item in the stack is a tuple of (indent_level, node)
    stack = [(0, hierarchy)]

    # Loop over each line in the input text
    for line in lines:
        
        # Skip empty lines
        if line.strip() == '':
            continue

        # Calculate the indentation level of the line (number of leading spaces divided by 4)
        indent_level = line.count('    ')
        
        # Get the name of the node from the line (strip leading and trailing spaces)
        name = line.strip()

        # Pop from the stack while the current indent_level is less than or equal to the indent_level of the top item on the stack
        while len(stack) > 1 and indent_level <= stack[-1][0]:
            stack.pop()

        # Add new node to the top node on the stack
        if stack:
            stack[-1][1][name] = {}

        # Add new node to stack
        stack.append((indent_level, stack[-1][1][name]))

    # After all lines have been processed, return the final hierarchy
    return hierarchy

if TESTING:
    jsonDict= text_to_json(text_input)
    print(jsonDict)
```

```python
def json_to_text(json_dict, indent=0):
    """
    Convert a JSON-like dictionary into indented text.

    Parameters:
    - json_dict (dict): The JSON-like dictionary to convert. Each key is a string, and its value is a
                        dictionary of its child nodes.
    - indent (int): The current indentation level. This parameter is used internally and should not
                    be set when calling this function. It defaults to 0.

    Returns:
    - str: The JSON-like dictionary converted to indented text.
    """
    
    # Create an empty string to build up the final text output
    text = ""

    # Loop over each key in the dictionary
    for key in json_dict:
        
        # Add to the text a line for the current key, indented according to the current indentation level
        text += '    '*indent + key + "\n"
        
        # Make a recursive call to process the child nodes of the current key
        # Pass the current indentation level plus one to the recursive call
        text += json_to_text(json_dict[key], indent=indent+1)
    
    # After all keys have been processed and all recursive calls have returned, return the final text
    return text

if TESTING:
    JtoT= json_to_text(jsonDict)
    print(JtoT)
```

```python
def json_to_dot(json_dict, parent_name=None):
    """
    Convert a JSON-like dictionary into DOT (graphviz) notation.

    Parameters:
    - json_dict (dict): The JSON-like dictionary to convert. Each key is a node, and its value is a
                        dictionary of its child nodes.
    - parent_name (str): The name of the parent node of the current node. This parameter is used
                         internally and should not be set when calling this function.

    Returns:
    - str: The JSON-like dictionary converted to DOT notation.
    """
    
    dot_string = ""

    # Loop over each key in the dictionary
    for key in json_dict:
        
        # If a parent name was given, add an edge from the parent to the current node to the DOT string
        if parent_name is not None:
            dot_string += f'"{parent_name}" -> "{key}";\n'
        else: # Node without a parent (root node)
            dot_string += f'"{key}";\n'
        
        # Make a recursive call to process the child nodes of the current key
        # The current key becomes the parent name for the recursive call
        dot_string += json_to_dot(json_dict[key], parent_name=key)

    # After all keys have been processed and all recursive calls have returned, return the DOT string
    return dot_string

if TESTING:
    dotOutput = json_to_dot(jsonDict)
    print(dotOutput)
```

```python
def wrap(dot_output, graph_type='digraph'):
    """
    Wrap a DOT notation string with the appropriate graph declaration.

    Parameters:
    - dot_output (str): The DOT notation string to wrap.
    - graph_type (str): The type of graph. This should be either 'digraph' for a directed graph or 'graph'
                        for an undirected graph. Defaults to 'digraph'.

    Returns:
    - str: The wrapped DOT notation string.
    """
    if graph_type == 'graph':
        dot_output = dot_output.replace('->', '--')
 
    return f'{graph_type} {{\n{dot_output}\n}}'

dot_output = """
"one";
"two";
"two" -> "sub1";
"two" -> "sub";
"sub" -> "2.3";
"2.3" -> "2.3.1";
"2.3" -> "2.3.2";
"three";
"""
if TESTING:
    wrapped_dot_output = wrap(dot_output)
    print(wrapped_dot_output)
```

```python
def TtoG(text, graph_type='digraph'):
    """
    Convert indented text into a graph using the graphviz Python library.

    Parameters:
    - text (str): The indented text to convert. Each line represents a node, and the indentation level
                  of the line represents the node's level in the hierarchy.
    - graph_type (str): The type of graph. This should be either 'digraph' for a directed graph or 'graph'
                        for an undirected graph. Defaults to 'digraph'.

    Returns:
    - Source: The rendered graph.
    """
    
    # Convert the text to a JSON-like dictionary
    json_dict = text_to_json(text)

    # Convert the JSON-like dictionary to DOT notation
    dot_output = json_to_dot(json_dict)

    # Wrap the DOT notation string with the appropriate graph declaration
    dot_output = wrap(dot_output, graph_type)

    # Create a Source object from the DOT string
    graph = Source(dot_output)
    
    graph.format = 'png'
    graph.render(filename='lastGraph', cleanup=True)

    
    # Return the graph
    return graph
if TESTING:
    graph = TtoG(text_input)
    graph
```
